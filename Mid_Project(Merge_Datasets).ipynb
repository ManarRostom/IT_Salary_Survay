<span style="color:#2929a3;font-size:15px;">Columns of <strong>df_2018 </strong> Dataset</span>

```python
df_2018_columns = ['Timestamp','Age','Gender','City','Position','Years of experience',
                   'Your level','Current Salary','nan','Salary one year ago',
                   'nan','Main language at work','nan','nan','nan',
                   'Company type','Company size'
                  ]
```

<span style="color:#2929a3;font-size:15px;">Columns of <strong>df_2019 </strong> Dataset</span>

```python
df_2019_columns = ['Zeitstempel','Age','Gender','City',
                   'Position (without seniority)','Years of experience',
                   'Seniority level','Yearly brutto salary (without bonus and stocks)',
                   'bonus + stocks' ,
                   'Yearly brutto salary (without bonus and stocks) one year ago. Only answer if staying in same country' ,
                   'bonus + stocks one year age' , 'Main language at work',
                   'Your main technology / programming language',
                   'Number of vacation days','Сontract duration','Company type',
                   'Company size'
                  ]
```

<span style="color:#2929a3;font-size:15px;">Columns of <strong>df_2020</strong> Dataset</span>

```python
df_2020_columns = ['Timestamp','Age','Gender','City','Position ',
                   'Total years of experience','Seniority level',
                   'Yearly brutto salary (without bonus and stocks) in EUR',
                   'Yearly bonus + stocks in EUR',
                   'Annual brutto salary (without bonus and stocks) one year ago. Only answer if staying in the same country',
                   'Annual bonus+stocks one year ago. Only answer if staying in same country',
                   'Main language at work',
                   'Your main technology / programming language',
                   'Number of vacation days','Сontract duration','Company type',
                   'Company size'
                  ]
```

<span style="color:#2929a3;font-size:15px;">Columns of <strong>Final</strong> Dataset</span>

```python
df_final_columns = ['Timestamp','Age','Gender','City',
                    'Position','Years_of_Experience','Level','Current_Salary',
                    'Bonus+Stocks','Salary_One_Year_Ago','Bonus+Stocks_One_year_Ago',
                    'Main_Language_at_Work','Main_Technology',
                    'Number_of_Vacation_Days','Contract_Duration','Company_Type',
                    'Company_Size'
                   ]
```

<span style="color:#2929a3;font-size:15px;">Merge Data Function</span>

```python
def Merge_Data(df_2018_columns,df_2019_columns,df_2020_columns,df_final_columns):
    df = pd.DataFrame()
    ## nan_series is a series of nan values used when a column doesn't exist in df_2018
    nan_series = pd.Series([np.nan for i in range(df_2018.shape[0])])

    for i in range(17):
        if df_2018_columns[i] == 'nan':
            if df_2019_columns[i] == 'bonus + stocks':
                df[df_final_columns[i]] = pd.concat([nan_series,(df_2019['Yearly bonus'] + df_2019['Yearly stocks']),df_2020[df_2020_columns[i]]])
            elif df_2019_columns[i] == 'bonus + stocks one year age':
                df[df_final_columns[i]] = pd.concat([nan_series,(df_2019['Yearly bonus one year ago. Only answer if staying in same country'] + df_2019['Yearly stocks one year ago. Only answer if staying in same country']),df_2020[df_2020_columns[i]]])
            else:
                df[df_final_columns[i]] = pd.concat([nan_series,df_2019[df_2019_columns[i]],df_2020[df_2020_columns[i]]])
        else:
            df[df_final_columns[i]] = pd.concat([df_2018[df_2018_columns[i]],df_2019[df_2019_columns[i]],df_2020[df_2020_columns[i]]])
    return df
```

<span style="color:#2929a3;font-size:15px;">Calling the Merge_Data Function</span>

```python
df = Merge_Data(df_2018_columns,df_2019_columns,df_2020_columns,df_final_columns)
```
